```python
import pandas as pd
import numpy as np
df = pd.read_csv('iris.csv')


from sklearn.preprocessing import LabelEncoder, OneHotEncoder
encoder = LabelEncoder()
df = df.drop(['Id'], axis=1)
df['Species']= encoder.fit_transform(df['Species'])
df
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)

df
X_scaled
```

```python
x_columns = 4
X = df.iloc[:, :x_columns].values
y = df.iloc[:, x_columns].values
X_scaled = scaler.fit_transform(X)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.20, random_state=7)

from sklearn.linear_model import LogisticRegression
l_model = LogisticRegression(random_state=7, solver='lbfgs', multi_class='ovr')
l_model.fit(X_train, y_train)

y_pred = l_model.predict(X_test)
y_pred
```

```python
from sklearn.neighbors import KNeighborsClassifier
K = 4
k_model = KNeighborsClassifier(n_neighbors=K)
k_model.fit(X_train, y_train)
y_pred = k_model.predict(X_test)
y_pred
k_score = k_model.score(X_test, y_test)
k_score
```

```python
from sklearn.svm import SVC
svm_model = SVC(max_iter = 100)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree_model = DecisionTreeClassifier(criterion = 'entropy', max_leaf_nodes = 50)
```

```python
from sklearn.ensemble import RandomForestClassifier
RF_model = RandomForestClassifier(n_estimators = 100, bootstrap = True)
```

```python
from sklearn.naive_bayes import GaussianNB
NB_model = GaussianNB()
```

```python
all_models = [l_model, k_model, svm_model, tree_model, RF_model, NB_model]
all_models
```

```python
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
for i in all_models:
    print(i)
    i.fit(X_train, y_train)
    y_pred = i.predict(X_test)
    print(y_pred)
    s = i.score(X_test, y_test)
    print(f'Accuracy: {s}')
    print(confusion_matrix(y_test, y_pred))
    #sn.pairplot(df, hue = "Species").set_title('Algoritm Comparison')
```
